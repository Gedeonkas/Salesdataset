```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Replace 'path/to/your/file.csv' with the actual path to your CSV file in Google Drive
path=r'/content/drive/MyDrive/SALES2019DATASET'
files = [file for file in os.listdir(path)]
print(files)
```

```python
salesdata=pd.DataFrame()
for file in files:
  current_data=pd.read_csv(path+"/"+file)
  salesdata=pd.concat([salesdata,current_data])
  print(salesdata)
```

```python
salesdata.to_csv(path+'salesdata.csv', index=False)
salesdata.dtypes
salesdata.describe()
salesdata.head()
salesdata
```

```python
salesdata.isnull().sum()
```

```python
salesdata=salesdata.dropna()
salesdata.isnull().sum()
salesdata
```

```python
def m(d):
  return d.split('/')[0]
salesdata['month']=salesdata['Order Date'].apply(m)
salesdata
```

```python
salesdata['month'].unique()
```

```python
salesdata=salesdata[salesdata['month']!='Order Date']
salesdata['month'].unique()
```

```python
salesdata.dtypes
```

```python
salesdata['month']= salesdata['month'].astype(int)
salesdata['Quantity Ordered']=salesdata['Quantity Ordered'].astype(int)
salesdata['Price Each']=salesdata['Price Each'].astype(float)
salesdata.dtypes
```

```python
salesdata['Sales']=salesdata['Quantity Ordered']*salesdata['Price Each']
salesdata
```

```python
salesdata.groupby('month')['Sales'].sum()
```

```python
months=range(1,13)
plt.bar(months,salesdata.groupby('month')['Sales'].sum())
plt.title('SALES BY MONTH',size=20)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()
```

```python
salesdata.groupby('Product')['Sales'].sum()
```

```python
salesdata.groupby('Product')['Sales'].sum().plot(kind='bar')
plt.title('SALES BY PRODUCT',size=20)
plt.xticks(rotation='vertical',size=6)
plt.ylabel('Sales in USD ($): Ech 1/1M')
plt.xlabel('Product name')
plt.show()
```

```python
def p(d):
  return d.split(',')[1]
salesdata['city']=salesdata['Purchase Address'].apply(p)
salesdata
```

```python
# prompt: le graphique de vente par ville
salesdata.groupby('city')['Sales'].sum().plot(kind='bar')
plt.title('SALES BY CITY',size=20)
plt.xticks(rotation='vertical',size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

```python
product=salesdata.groupby('Product')['Quantity Ordered'].sum().index
quantity=salesdata.groupby('Product')['Quantity Ordered'].sum()
price=salesdata.groupby('Product')['Price Each'].mean()
plt.figure(figsize=(10,5))
fig, ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(product,quantity,color='g')
ax2.plot(product,price,'b-')
ax1.set_xticks(product)
ax1.set_xticklabels(product,rotation='vertical',size=8  )
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered',color='g')
ax2.set_ylabel('Price ($)',color='g')
plt.title('Oders vs Price Terendline',size=40, color='b')
plt.show()
```
